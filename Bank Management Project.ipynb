```python
import pandas
import csv
print("\n\nWelcome to Bank Management System")
def Main():
    while 1:
        print("************************************************")
        print("==========1.open new client account===========")
        print("==========2.All Registerd Accounts =========")
        print("==========3.Quit==============================")
        print("************************************************")
        a = int(input("Enter the number from main menu to proceed: "))
        if a==1:
            print("==========Open new client account===========")
            New_Client()
        elif a==2:
            Pre()
        else:
            quit()
        return
def New_Client():
    while 1:
        print("\topening new account")
        print("1.salary account")
        print("2.Savings account")
        print("3.Main Menu")
        b=int(input("Enter a number from above menu"))
        if b==1:
            print("Welcome to Salary account")
            Salary()
        elif b==2:
            print("Welcome to Saving Account")
            Saving()
        else:
            quit()
        return
def Pre():
    while 1:
        print("Welcome to all registered account Details :")
        print("1.Show Previous account details: ")
        print("2.Delete Previous account details: ")
        print("3.updte function:")
        print("4.Exit")
        c = int(input("Enter the number from menu you want to proceed: "))
        if c==1:
            fd = pandas.read_csv('Niaz.csv')
            print(fd)
        
        elif c==2:
            n = int(input("Enter the Serial number whose data you want to delete:"))
            v = pandas.read_csv('Niaz.csv')
            d = v[: 6]
            d.set_index('Sr', inplace=True)
            d = d.drop(n)
            print(d)
        elif c==3:
            print("Update function")
            n=int(input("Enter row no:"))
            N=str(input("Enter col no:"))
            r=str(input("enter new thing:"))
            v=n-1
            df=pandas.read_csv("Niaz.csv")
            df.loc[v,N]=r
            df.to_csv("Adeel.CSV",index=False)
            print(df)
        else:
            quit()
        return
def Salary():
    from csv import writer
    N = str(input("Enter the name: "))
    F = str(input("Enter the Father name: "))
    CNIC = int(input("Enter CNIC Number: "))
    A = str(input("Enter your Adress: "))
    M = str(input("Enter your mobile number: "))
    list_data = [N, F, CNIC, A, M, 'Salary']
    with open('Adeel.CSV', 'a', newline='') as f_object:
        writer_object= writer(f_object)
        writer_object.writerow(list_data)
        f_object.close()
        print("Your Account Created Succesfully")
        print("Enter 0 for Main Menu")
        print("Enter 1 for previous Menu")
        d = int(input("Enter from Menu to Proceed"))
        if d == 0:
            Main()
        elif d == 1:
            New_Client()
        else:
            print("Invalid Entry")
            
def Saving():
    from csv import writer
    N = str(input("Enter the name: "))
    F = str(input("Enter the Father name: "))
    CNIC = int(input("Enter CNIC Number: "))
    A = str(input("Enter your Adress: "))
    M = str(input("Enter your mobile number: "))
    list_data = [N, F, CNIC, A, M, 'Saving']
    with open('Adeel.CSV', 'a', newline='') as f_object:
        writer_object= writer(f_object)
        writer_object.writerow(list_data)
        f_object.close()
        print("Your Account Created Succesfully")
        print("Enter 0 for Main Menu")
        print("Enter 1 for previous Menu")
        e = int(input("Enter from Menu to Proceed"))
        if e == 0:
            Main()
        elif e == 1:
            New_Client()
        else:
            print("Invalid Entry")
Main()
```
